```python
import pandas as pd
```

```python
eda_df=pd.read_csv('../../data/raw/incident_event_log.csv')
```

```python
eda_df['category'].unique()
```

```python
category_details = (
    eda_df.groupby('category')['made_sla']
    .agg(
        Total_incidents='count',
        Breaches=lambda x: (x == False).sum()
    )
    .reset_index()
)

category_details['Breach_percentage'] = (
    category_details['Breaches']
    / category_details['Total_incidents']
) * 100

category_details = category_details.sort_values(
    by='Breach_percentage',
    ascending=False
)
```

```python
category_details
%store category_details
```

```python
subcategory_details=eda_df.groupby('subcategory')['made_sla'].agg(Total_incidents='count',Breaches=lambda x:(x==False).sum()).reset_index()

subcategory_details['Breach_percentage']=((subcategory_details['Breaches']/subcategory_details['Total_incidents'])*100).round(2)

subcategory_details.sort_values(by='Breaches',ascending=False)

```

```python
from datetime import datetime
```

```python
eda_df['Hour'] = pd.to_datetime(eda_df['opened_at'], format="%d/%m/%Y %H:%M").dt.hour
eda_df['Day_of_week']=pd.to_datetime(eda_df['opened_at'],format="%d/%m/%Y %H:%M").dt.day_name()
eda_df['Month']=pd.to_datetime(eda_df['opened_at'],format="%d/%m/%Y %H:%M").dt.month_name()
```

```python
opened_at_details_hour=eda_df.groupby('Hour')['made_sla'].agg(Total_incidents='count',Breaches=lambda x:(~x).sum())
opened_at_details_hour['Breach_rate']=((opened_at_details_hour['Breaches']/opened_at_details_hour['Total_incidents'])*100).round(2)
opened_at_details_hour
%store opened_at_details_hour
```

```python
opened_at_details_day_of_week=eda_df.groupby('Day_of_week')['made_sla'].agg(Total_incidents='count',Breaches=lambda x:(~x).sum())
opened_at_details_day_of_week['Breach_rate']=((opened_at_details_day_of_week['Breaches']/opened_at_details_day_of_week['Total_incidents'])*100).round(2)
opened_at_details_day_of_week
```

```python
%store opened_at_details_day_of_week
```

```python
opened_at_details_month=eda_df.groupby('Month')['made_sla'].agg(Total_incidents='count',Breaches=lambda x:(~x).sum())
opened_at_details_month['Breach_rate']=((opened_at_details_month['Breaches']/opened_at_details_month['Total_incidents'])*100).round(2)
opened_at_details_month
```

```python
%store opened_at_details_month
```

```python
opened_at_details_hour[(opened_at_details_hour['Total_incidents']>=500) & (opened_at_details_hour['Breach_rate']>=6.5)].sort_values(by='Breach_rate',ascending=False)
```

```python
opened_at_details_day_of_week[(opened_at_details_day_of_week['Total_incidents']>=500) & (opened_at_details_day_of_week['Breach_rate']>=6.5)].sort_values(by='Breach_rate',ascending=False)
```

```python
opened_at_details_month[(opened_at_details_month['Total_incidents']>=500) & (opened_at_details_month['Breach_rate']>=6.5)].sort_values(by='Breach_rate',ascending=False)
```

```python
time=pd.concat([opened_at_details_hour,opened_at_details_day_of_week,opened_at_details_month],keys=['opened_at_details_hour','opened_at_details_day_of_week','opened_at_details_month'])
```

```python
time[(time['Total_incidents']>=500) & (time['Breach_rate']>=6.5)].sort_values(by='Breach_rate',ascending=False).sort_index().reset_index()
```
