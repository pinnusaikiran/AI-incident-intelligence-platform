## Target Variable Transformation (Business-Oriented Target Encoding)

The original target variable (`made_sla`) is represented as:

- **True** → Incident met the SLA
- **False** → Incident breached the SLA

However, the primary business objective of this project is **to identify incidents that are likely to breach the SLA before they occur**, enabling proactive intervention by the operations team.

Most machine learning evaluation metrics in Scikit-Learn, such as **Precision, Recall, F1-Score, ROC-AUC**, along with model selection techniques like **Cross-Validation, GridSearchCV, RandomizedSearchCV, and Optuna**, consider the class labeled **1** as the positive class by default.

To align the machine learning objective with the business objective, the target variable is re-encoded as:

| Original Value | New Value | Meaning |
|---------------|-----------|---------|
| True | 0 | SLA Met |
| False | 1 | SLA Breach |

This transformation allows all evaluation metrics and hyperparameter optimization procedures to naturally optimize for **SLA breach prediction**, eliminating the need for custom scoring functions (`pos_label=0`) and making the modeling pipeline cleaner, more interpretable, and consistent with the business goal.

```python
y=feature_df['made_sla'].map({True:0,False:1})

# Alternative implementation
# y=(~feature_df['made_sla']).astype(int)
y.shape
print(y.value_counts())
```

## Missing Value Analysis (Training Dataset)

Missing value analysis was performed exclusively on the training dataset to avoid introducing information from the test dataset into feature engineering decisions.

The symbol '?' is treated as an unknown category rather than a traditional missing value because it may carry business meaning in the incident management process.

```python
X_train_missing=(X_train[['contact_type','location','category','subcategory','u_symptom','assignment_group','assigned_to','caller_id','opened_by']]=='?').sum()
X_train_missing=X_train_missing.reset_index()
X_train_missing.columns=['features','Missing Count (?)']
X_train_missing['Missing %']=((X_train_missing['Missing Count (?)']/len(X_train))*100).round(2)
X_train_missing
```

## Ordinal Feature Engineering

Impact, urgency and priority contain an inherent business hierarchy.

Unlike nominal categorical variables, the order between categories carries meaningful information.

Therefore, ordinal encoding was selected to preserve the severity relationship while converting categories into numerical representations suitable for machine learning models.

### Ordinal Mapping Strategy

Impact:
- Low → 1
- Medium → 2
- High → 3

Urgency:
- Low → 1
- Medium → 2
- High → 3

Priority:
- Low → 1
- Moderate → 2
- High → 3
- Critical → 4

This mapping ensures that higher numerical values correspond to higher business severity, improving interpretability during model evaluation and SHAP analysis.

## Data Leakage Prevention

To ensure realistic model evaluation:

1. Feature selection was completed before modeling.
2. Train-test split was performed before any encoding.
3. Frequency encoding mappings will be learned from training data only.
4. Target encoding mappings will be learned from training data only.
5. Unseen categories in test/production data will use the global target mean.
6. SMOTE (if used) will be applied only on training data.
7. Calendar features (Month and Day_of_week) will be treated as categorical variables and transformed within the training pipeline.

## Feature Engineering Conclusion
Feature engineering decisions were finalized based on business understanding, EDA findings, cardinality analysis, missing value analysis and leakage prevention principles. The resulting feature set will be transformed through a preprocessing pipeline in the modeling phase. The implementation of preprocessing, encoding, imbalance handling and model training will be performed in 05_model_training.ipynb using Scikit-Learn Pipelines and ColumnTransformer.
All learned transformations such as frequency encoding and target encoding will be fitted exclusively on the training dataset to avoid data leakage.
